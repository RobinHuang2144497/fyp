```python
import requests
url = 'https://guba.eastmoney.com/o/list,sszh000001.html'
r = requests.get(url)

from bs4 import BeautifulSoup
items = []  # 定义一个空列表，存储帖子信息
soup = BeautifulSoup(r.text, 'html.parser')
post_list = soup.find_all('div', class_='articleleh_normal_post')  # 找到所有的帖子所在标签
for post in post_list:
    read_counts = post.find('span', class_='11 al').text  # 获取帖子阅读数
    comment_counts = post.find('span', class_='12 a2').text  # 获取帖子评论数
    title = post.find('span', class_='13 a3').text  # 获取帖子标题
    author_id = post.find('span', class_='14 a4').text  # 获取作者id
    date = post.find('span', class_='15 a5').text  # 获取更新时间
    item = [read_counts, comment_counts, title, author_id, date]
    items.append(item)

import pandas as pd
df = pd.DataFrame(items, columns=['阅读数', '评论数', '标题', '作者', '最后更新'])
df.to_excel('帖子.xlsx')
```
